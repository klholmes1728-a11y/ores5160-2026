Q3. Data Visualization

```python
airquality["pollutant"].unique()
```

```python
airquality = pd.read_csv("airquality.csv")
```

```python
airquality["pollutant"].unique()
```

```python
pm25_data = airquality[airquality["pollutant"] == "pm25"]
```

```python
import matplotlib.pyplot as plt

pm25_data["value"].hist(bins=20)
plt.xlabel("PM2.5 Value")
plt.ylabel("Frequency")
plt.title("Histogram of PM2.5 Values")
plt.show()
```

```python
pm25_data = airquality[airquality["pollutant"] == "pm25"]

import matplotlib.pyplot as plt

plt.boxplot(pm25_data["value"])
plt.ylabel("PM2.5 Value")
plt.title("Boxplot of PM2.5 Values")
plt.show()
```
